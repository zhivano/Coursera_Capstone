```python
#Importing libraries
import numpy as np  
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)


import json
!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim

import requests
from pandas.io.json import json_normalize

import matplotlib.cm as cm
import matplotlib.colors as colors

from sklearn.cluster import KMeans
!conda install -c conda-forge folium=0.5.0 --yes 
import folium

print('Libraries imported.')
```

```python
#Importing BeautifulSoup
from bs4 import BeautifulSoup
import lxml
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
#Transforming a html table into a dataframe
df= pd.read_html(tab)
df_list=df[0]
df_list.head(102)
```

```python
#Dropping rows where the borough is 'Not assigned'
df_l=df_list[df_list.Borough!='Not assigned']
df_l.head()
```

```python
#Grouping rows by postal code and borough and combining neighbourhoods with the same postal code
df_l1 = df_l.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df_l1.reset_index(inplace=True)
df_l1.head(150)
```

```python
#Replacing "Not Assigned" neighbourhood with the respective borough
mask = df_l1.Neighbourhood=='Not Assigned'
column_name = 'Neighbourhood'
df_l1.loc[mask, column_name] =df_l1['Borough']
df_l1.head(30)
```

```python
#Checking for any cell "Not Assigned" left in the column "Neighbourhood"
df_l1['Neighbourhood'].str.contains('Not Assigned')
```

```python
coordinates=pd.read_csv('http://cocl.us/Geospatial_data')
coordinates.head()
```

```python
df_coordinates=pd.merge(df_l1,coordinates,on='Postal Code')
df_coordinates.head()
```

```python
#Clustering all boroughs containing 'Toronto'
df_cluster=df_coordinates[df_coordinates['Borough'].str.contains('Toronto', regex=False)]
df_cluster
```

```python
#Mapping Toronto
mapt = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

for lat,lng,borough,neighbourhood in zip(df_cluster['Latitude'],df_cluster['Longitude'],df_cluster['Borough'],df_cluster['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(mapt)
mapt
```

```python
k=5
toronto_clustering = df_cluster.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df_cluster.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

#Inserting colours & pop-s
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df_cluster['Latitude'], df_cluster['Longitude'], df_cluster['Neighbourhood'], df_cluster['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
